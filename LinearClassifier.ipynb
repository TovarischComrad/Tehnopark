```python
import pandas as pd

df = pd.read_csv('results.csv')
df
```

```python
df.columns
```

```python
lst = ['Hindi', 'English', 'Science', 'Maths', 'History',
       'Geograpgy', 'Results']
```

```python
df = df[lst]
df
```

```python
df.isnull().sum()
```

```python
X = df[lst[:-1]]
X
```

```python
y = df['Results']
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression().fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
X_test
```

```python
y_pred
```

$accuracy = \frac{TP + }{TP + TN + FP + FN}$

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

$precision = \frac{TP}{TP + FP}$

```python
from sklearn.metrics import precision_score

precision_score(y_test, y_pred)
```

$recall = \frac{TP}{TP + FN}$

```python
from sklearn.metrics import recall_score

recall_score(y_test, y_pred)
```

$F = \frac{2\cdot precision \cdot recall}{precision + recall}$

```python
from sklearn.metrics import f1_score

f1_score(y_test, y_pred)
```

```python
def mark_predict(hindi, english, science, maths, history, geography):
  marks = [[hindi, english, science, maths, history, geography]]
  df2 = pd.DataFrame(marks, columns=lst[:-1])
  result = clf.predict(df2)[0]
  return result
```
