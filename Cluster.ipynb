```python
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs

# data = [[x1, y1], [x2, y2], ..., [xn, yn]]
# labels = [c1, c2, ..., cn]

N = 50
data, labels = make_blobs(n_samples=N, centers=2, random_state=0)
print(data[:10])
print(labels[:10])
```

```python
def plot(data, labels):
    x1 = data[labels==0, 0]
    x2 = data[labels==1, 0]
    y1 = data[labels==0, 1]
    y2 = data[labels==1, 1]
    plt.scatter(x1, y1, color='red')
    plt.scatter(x2, y2, color='green')
```

```python
def plot_c(data, labels, xc1, yc1, xc2,  yc2):
    x1 = data[labels==0, 0]
    x2 = data[labels==1, 0]
    y1 = data[labels==0, 1]
    y2 = data[labels==1, 1]
    plt.scatter(x1, y1, color='lightcoral')
    plt.scatter(x2, y2, color='lime')
    plt.scatter(xc1, yc1, color='red')
    plt.scatter(xc2, yc2, color='green')
```

```python
plot(data, labels)
```

```python
import numpy as np

clusters = np.zeros(N)
clusters
```

```python
plot(data, clusters)
```

```python
plot_c(data, clusters, 1, 2, 3, 4)
```

```python
import random

i1 = random.randint(1, N)
i2 = random.randint(1, N)

xc1, yc1 = data[i1][0], data[i1][1]
xc2, yc2 = data[i2][0], data[i2][1]
```

```python
plot_c(data, clusters, xc1, yc1, xc2, yc2)
```

```python
import math

def dist(x1, y1, x2, y2):
  return math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
```

```python
dist(1, 2, 3, 4)
```

```python
def new_clusters(data, clusters, xc1, yc1, xc2, yc2):
  for i in range(len(data)):
    p = data[i]
    d1 = dist(p[0], p[1], xc1, yc1)
    d2 = dist(p[0], p[1], xc2, yc2)
    if d1 < d2:
      clusters[i] = 0
    else:
      clusters[i] = 1
  return clusters
```

```python
cl2 = new_clusters(data, clusters, xc1, yc1, xc2, yc2)
plot_c(data, cl2, xc1, yc1, xc2, yc2)
```

```python
def update_centers(data, clusters):
      x1 = data[clusters==0, 0]
      x2 = data[clusters==1, 0]
      y1 = data[clusters==0, 1]
      y2 = data[clusters==1, 1]
      xc1 = np.average(x1)
      xc2 = np.average(x2)
      yc1 = np.average(y1)
      yc2 = np.average(y2)
      return xc1, yc1, xc2, yc2
```

```python
a, b, c, d = update_centers(data, cl2)
plot_c(data, cl2, a,  b, c, d)
```

```python
def clusterization(data, max_iter):
    N = len(data)
    clusters = np.zeros(N)

    #  1.   
    i1 = random.randint(1, N)
    i2 = random.randint(1, N)
    xc1, yc1 = data[i1][0], data[i1][1]
    xc2, yc2 = data[i2][0], data[i2][1]

    #  2.  
    for i in range(max_iter):

      #  2.1   
      clusters = new_clusters(data, clusters, xc1, yc1, xc2, yc2)

      #  2.2   
      xc1, yc1, xc2, yc2 = update_centers(data, clusters)

    return clusters
```

```python
result = clusterization(data, 50)
plot(data, result)
```
