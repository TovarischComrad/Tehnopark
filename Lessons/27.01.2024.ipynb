```python
import pandas as pd

df = pd.read_csv('data.csv')
df
```

```python
df.columns
```

```python
lst = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
       'floors', 'waterfront', 'view', 'condition', 'sqft_above',
       'sqft_basement', 'yr_built', 'city',
       ]
df = df[lst]
df
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
encoded_data = label_encoder.fit_transform(df['city'])
df['city'] = encoded_data
df
```

```python
encoded_data = label_encoder.fit_transform(df['country'])
df['country'] = encoded_data
df
```

```python
df
```

```python
lst = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
       'floors', 'waterfront', 'view', 'condition', 'sqft_above',
       'sqft_basement', 'yr_built', 'city',
       ]
```

```python
y = df['price']
X = df[lst]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y_test, y_pred)
```

```python
import math
from sklearn.metrics import mean_squared_error
math.sqrt(mean_squared_error(y_test, y_pred))
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test, y_pred)
```
